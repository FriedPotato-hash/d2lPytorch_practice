# 7.7.1. From ResNet to DenseNet

Recall the taylor expansion for functions. 



함수를 higher order의 항으로 decompose 할 수 있다는 것이 포인트.
ResNet은 다음과 같이 함수를 분해한다.

### 𝑓(𝐱)=𝐱+𝑔(𝐱).

즉, 선형함수와 비선형함수의 조합으로 분해하는 것.
두개의 항 이상의 정보를 얻고 싶다면 어떻게 할 것인가?라는 고민에 나온 것이 

DenseNet이다.



ResNet과 DenseNet의 차이점은, 전자는 'added' 되지만, 후자의 경우 output이 concaternated 된다는 점이다. 

그 결과로 x를 매우 복잡한 sequence of function으로 mapping 하는 것이 가능해졌다. 
다음은 그 결과.

### 𝐱→[𝐱,𝑓1(𝐱),𝑓2([𝐱,𝑓1(𝐱)]),𝑓3([𝐱,𝑓1(𝐱),𝑓2([𝐱,𝑓1(𝐱)])]),…]

DenseNet은, 변수들간의 dependency graph가 매우 dense하다는 것에서 비롯된 이름이다.
chain의 마지막 layer는 이전 모든 layer들에 대해서 densely connected 되어 있다.



DenseNet을 구성하는 주요한 요소들은 *dense blocks*와 *transition layers*인데, 전자는 input과 output이 어떻게 concatenated되는지 정의하고, 후자는 channel의 개수를 너무 크지 않게 조절한다.

# 7.7.2. Dense Blocks

```python
# DenseNet uses modified "batch normalization, activation, convolution" structure of ResNet.

import torch
from torch import nn
from d2l import torch as d2l

def conv_block(input_channels, num_channels):
    return nn.Sequential(
        nn.BatchNorm2d(input_channels), nn.ReLU(),
        nn.Conv2d(input_channels, num_channels, kernel_size=3, padding=1))

# Dense Block은 여러개의 convolution block들로 구성된다. 
# 이때 convolution layer들은 각각 같은 수의 output channel을 갖는다.
# 하지만, forward propagation에서 각 convolution block의 input과 Output은 channel dimension에서 concatenate된다.
class DenseBlock(nn.Module):
    def __init__(self, num_convs, input_channels, num_channels):
        super(DenseBlock, self).__init__()
        layer = []
        for i in range(num_convs):
            layer.append(conv_block(
            num_channels*i + input_channels, num_channels))
        self.net = nn.Sequential(*layer)
        
    def forward(self,X):
        for blk in self.net:
            Y = blk(X)
            # 각 block의 input과 output을 channel dim에서 concat한다.
            X = torch.cat((X,Y), dim=1)
        return X
    
    
blk = DenseBlock(2,3,10)
X = torch.randn(4,3,8,8)
Y = blk(X)

Y.shape
```

# 7.7.3. Transition Layers

각각의 dense layer가 channel의 수를 증가시키는 것을 위에서 확인했을 것.
그러나, 알다 싶이 너무 많은 channel의 개수는 지나치게 복잡한 모델을 만들 수 있다.

따라서 transition layer는 이러한 complexity를 조정하는 역할을 한다.

1. 1x1 conv layer를 사용하여, channel의 개수를 감소시키며, 
2. average pooling(stride=2)을 통해 height, width를 절반씩으로 감소시킨다.

```python
def transition_block(input_channels, num_channels):
    return nn.Sequential(
        nn.BatchNorm2d(input_channels), nn.ReLU(),
        nn.Conv2d(input_channels, num_channels, kernel_size=1),
        nn.AvgPool2d(kernel_size=2, stride=2))

# 23개의 channel dim을 10으로 조정. & width,height은 절반으로
blk = transition_block(23,10)
blk(Y).shape
```

# 7.7.4. DenseNet Model

```python
b1 = nn.Sequential(
    nn.Conv2d(1,64,kernel_size=7, stride=2, padding=3),
    nn.BatchNorm2d(64), nn.ReLU(),
    nn.MaxPool2d(kernel_size=3, stride=2, padding=1))
```

```python
# num_channels: 현재 채널의 number
# growth_rate: 하나의 dense block을 지날 때마다, 출력으로 나오는 channel dim
num_channels, growth_rate = 64, 32
num_convs_in_dense_blocks = [4,4,4,4]
blks = []

for i, num_convs in enumerate(num_convs_in_dense_blocks):
    blks.append(DenseBlock(num_convs, num_channels, growth_rate))
    
    # previous dense block의 num of output channels
    num_channels += num_convs*growth_rate
    
    # Transition layer는 dense blocks들 사이에 channel 개수를 1/2 해줌
    if i != len(num_convs_in_dense_blocks) - 1:
        blks.append(transition_block(num_channels, num_channels//2))
        num_channels = num_channels//2
    
    
net = nn.Sequential(
    b1, *blks,
    nn.BatchNorm2d(num_channels), nn.ReLU(),
    nn.AdaptiveMaxPool2d((1,1)),
    nn.Flatten(),
    nn.Linear(num_channels, 10))
```

# 7.7.5. Training

```python
lr, num_epochs, batch_size = 0.1, 10, 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size, resize=96)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```

# 7.7.6. Summary

- cross-layer connection을 다루는 방법으로는, 단순 add up 해주는 ResNet과는 다르게 concatenates input & output 해주는 DenseLayer가 있다.
- DenseNet은 dense block & transition layer로 구성된다.
- 차원의 수를 다시 줄이는 방향으로 조정해 줄 필요가 있다.
